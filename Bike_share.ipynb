```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '/content/drive/MyDrive/ML Projects/Project_1/Reg_BikeShareDay.csv'
data = pd.read_csv(path)
data.head()
```

```python
# data.describe()
```

```python
data1 = data.copy()
data1 = data1.drop(['dteday' , 'instant' , 'casual' , 'registered'] , axis = 'columns')
data1.head()
```

```python
y = data1['cnt']
data2 = data1.drop(['cnt'] , axis = 'columns')
data2.head()
```

```python
len(data2.atemp)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(data2 , y , test_size= 0.3 ,random_state=10)
```

```python
from sklearn.svm import SVR
scalar = SVR(kernel='linear')
scalar.fit(x_train , y_train)
```

```python
scalar.score(x_test,y_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train , y_train)
```

```python
model.score(x_test,y_test)
```
